```python
from pathlib import Path
from collections import Counter
import pandas as pd

from battery_aging.data_preprocessing import load_dataset


Dataset = "Stanford"
data_folder = Path("..", "Battery_life_Dataset", Dataset)

datasets = load_dataset(
    data_folder,
    n_files="all",
    seed=42
)

protocols = [
    str(d["discharge_protocol"])
    for d in datasets
]

stanford_protocols = pd.DataFrame(
    Counter(protocols).items(),
    columns=["Discharge Protocol", "Frequency"]
)

stanford_protocols


```

```python
Dataset = "HUST"
data_folder = Path("..", "Battery_life_Dataset", Dataset)

datasets = load_dataset(
    data_folder,
    n_files="all",
    seed=42
)

protocols = [
    str(d["discharge_protocol"])
    for d in datasets
]

hust_protocols = pd.DataFrame(
    Counter(protocols).items(),
    columns=["Discharge Protocol", "Frequency"]
)

hust_protocols
```

```python
Dataset = "Tongji"
data_folder = Path("..", "Battery_life_Dataset", Dataset)

datasets = load_dataset(
    data_folder,
    n_files="all",
    seed=42
)

protocols = [
    str(d["discharge_protocol"])
    for d in datasets
]

tongji_protocols = pd.DataFrame(
    Counter(protocols).items(),
    columns=["Discharge Protocol", "Frequency"]
)

tongji_protocols
```

```python
protocol_comparison = pd.concat(
    [
        stanford_protocols.assign(Dataset="Stanford"),
        hust_protocols.assign(Dataset="HUST"),
        tongji_protocols.assign(Dataset="Tongji")
    ],
    ignore_index=True
)

protocol_comparison[
    ["Dataset", "Discharge Protocol", "Frequency"]
]
```
