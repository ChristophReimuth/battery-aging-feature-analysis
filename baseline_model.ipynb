```python
import pickle
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
with open("processed_battery_features.pkl", "rb") as f:
    feature_df = pickle.load(f)

feature_df = pd.DataFrame(feature_df)  # optional, falls nötig
```

```python
model_df = feature_df.dropna().copy()

features = model_df[["DTW_I", "DTW_V"]]
target = model_df["SOH"]

train_df, test_df = train_test_split(
    model_df,
    test_size=0.2,
    shuffle=False  # wichtig für Zeitreihe
)

features_train = train_df[["DTW_I", "DTW_V"]]
target_train = train_df["SOH"]

features_test = test_df[["DTW_I", "DTW_V"]]
target_test = test_df["SOH"]
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

model.fit(features_train, target_train)

pred_test = model.predict(features_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

print("MAE:", mean_absolute_error(target_test, pred_test))
print("R2:", r2_score(target_test, pred_test))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

plt.scatter(target_test, pred_test, alpha=0.6)

plt.xlabel("True SOH")
plt.ylabel("Predicted SOH")
plt.title("Baseline Model (Loaded Data)")
plt.grid(True)

plt.show()
```
